```python
import pandas as pd
import numpy as np
```

```python
health_full = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Health_Full.csv')
```

```python
qol_cols = {'Average Number of Physically Unhealthy Days' : .2, 
            'Average Number of Mentally Unhealthy Days' : .2,
           '% Insufficient Sleep': .2, 
            '% With Access to Exercise Opportunities': .4}

pop_cols = {'% Fair or Poor Health': .22, 
            '% Low Birthweight' : .14, 
            'Teen Birth Rate' : .08, 
            '% Uninsured' : .18, 
            'Life Expectancy': .22,
           'Age-Adjusted Death Rate': .16}

diet_cols = {'% Adults with Obesity': .16, 
             'Food Environment Index': .25, 
             '% Adults with Diabetes': .16, 
            '% Food Insecure': .21, 
             '% Limited Access to Healthy Foods': .22}
```

```python
qol = sum([health_full[col].dropna() * qol_cols[col] for col in qol_cols]) * .26
pop = sum([health_full[col].dropna() * pop_cols[col] for col in pop_cols]) * .37
diet = sum([health_full[col].dropna() * diet_cols[col] for col in diet_cols]) * .37
```

```python
health_full['health_raw'] =  qol.fillna(0) + pop.fillna(0) + diet.fillna(0)
```

```python
health_full.sort_values(by=['health_raw'])
```
