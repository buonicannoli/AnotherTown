```python
import numpy as np
import pandas as pd
```

```python
cols = ['CensusTract', 'State', 'County', 'POP2010', 'Urban', 'MedianFamilyIncome', 'PovertyRate', 'LowIncomeTracts',
        'LILATracts_1And10', 'LILATracts_halfAnd10','TractSNAP', 'TractLOWI', 'LALOWI1_20']
df = pd.read_csv('../Data/Health/food_access.csv', usecols = cols)
df
```

```python
result = []
for county in df['County'].unique():
    rows = df[df['County'] == county]
    result.append([rows.iloc[0,1], county] + ([ np.average(rows[col], weights=rows['POP2010']) for col in df.columns[3:]]))
    
result
```

```python
df_avg = pd.DataFrame(result, columns=df.columns[1:])
df_avg
```
