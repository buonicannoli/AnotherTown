```python
import pandas as pd
import numpy as np
from scipy.stats import binned_statistic
```

```python
df = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Demo_Raw.csv')
age = pd.read_csv('../Data/Politics/formatted-usa-2016-presidential-election-by-county.csv', 
                  usecols = ['State', 'ST', 'Fips', 'County', 'Median Age'])
df = df.drop(df[df['County'].isna()].index, axis=0)
df['% 18 to 65'] = 100 - (df['% less than 18 years of age'] + df['% 65 and over'])
df.head()
```

```python
for c, col in enumerate(df.columns[4:]):
    df[col] = pd.qcut(df[col].rank(method='first'), 10, labels=[i for i in range(1,11)]) 
bins = [0,1000,10000,20000,50000,100000,200000,300000,800000,2000000,20000000]
labels = list(range(1,11))
df['Population_bins'] = pd.cut(df['Population'], bins, labels=labels)
df.head()
```

```python
df['Diversity'] = (df['% Black'].astype('float64').fillna(0) + 
                   df['% Asian'].astype('float64').fillna(0) + 
                   df['% Hispanic'].astype('float64').fillna(0)
                  ) / 4
```
