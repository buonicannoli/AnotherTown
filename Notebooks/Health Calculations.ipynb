```python
import pandas as pd
import numpy as np
```

```python
health_full = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Health_Full.csv')
```

```python
qol_cols = ['Average Number of Physically Unhealthy Days', 
            'Average Number of Mentally Unhealthy Days',
           '% Insufficient Sleep', 
            '% With Access to Exercise Opportunities']

pop_cols = ['% Fair or Poor Health', 
            '% Low Birthweight', 
            'Teen Birth Rate', 
            '% Uninsured', 
            'Life Expectancy',
           'Age-Adjusted Death Rate']

diet_cols = ['% Adults with Obesity', 
             'Food Environment Index', 
             '% Adults with Diabetes', 
            '% Food Insecure', 
             '% Limited Access to Healthy Foods']
```

```python
qol_df = health_full[qol_cols]
pop_df = health_full[pop_cols]
diet_df= health_full[diet_cols]
dfs = [qol_df, pop_df, diet_df]
```

```python
weights_raw = [[2,2,2,6],[15,5,5,8,15,6],[2,6,2,4,4]]
weights = [[i/sum(weights) for i in weights] for weights in weights_raw]
df_str = ['qol', 'pop', 'diet']
cat_weights = [.26,.37,.27]

def zero_weight(zero_idx, w):
    w = [0  if idx + 3 in zero_idx else weight for idx, weight in enumerate(w)]
    return np.divide(w, sum(w))
    
#re-weight if nan in row
for idx, df in enumerate(dfs):
    temp = []
    nans = pd.isna(df).any(axis=1)
    for i, row in df.iterrows():
        w = weights[idx]
        if nans[i]:
            w = zero_weight([c for c, i in enumerate(row) if c > 2 and not i >= 0], w)
        temp.append(sum(list((row * w).fillna(0))) * cat_weights[idx])

    health_full[df_str[idx] + '_avg'] = temp
```

```python
health_full['health_raw'] = sum([health_full[col+ '_avg'] for col in df_str])
```
