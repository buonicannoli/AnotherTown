```python
safety = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Safety_Scores.csv', index_col=0)
health = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Health_Scores.csv', index_col=0)
demo = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Demo_Scores.csv', index_col=0)
economy =pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Economy_Scores.csv', index_col=0)
col = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/COL_Scores.csv', index_col=0)
political = pd.read_csv('C:/Users/User/Documents/GitHub/AnotherTown/Data/Cleaned/Political_raw.csv', index_col=0)
```

```python
df = demo.copy(deep=True,)
df= df.merge(safety, on=['FIPS', 'State', 'County'], how='outer')
df= df.merge(health, on=['FIPS', 'State', 'County'])
df=df.merge(economy, on=['FIPS', 'State', 'County'])
df=df.merge(col, on=['FIPS', 'State', 'County'])
df=df.merge(political, on=['FIPS'])
df.head(3)
df['Total'] = df.iloc[:,7:].mean(axis = 1, skipna = True) 
# df[df['Population_bins'] > 8].sort_values(by=['Total'])
```

```python
# weights = {'Safety_avg': 5, 'health_raw': 3, 'Economy Score': 8, 'col_scores': 8}
# df['Total'] = [mean([row[col]]*weights[col])  for i, row in df.iterrows() for col in list(weights.keys())]
df.head(3)
```

```python
#  [np.mean([[row[col]]*weights[col] for col in list(weights.keys())]) for i, row in df.iterrows() ]
# {'Safety_avg': 5, 'health_raw': 3, 'Economy Score': 8, 'col_scores': 8}
from numpy import ma
w = np.array([5,3,8,15,20])
result = []
for i, row in df.iloc[:,7:-1].iterrows():
    row = np.ma.MaskedArray(row, mask=np.isnan(row))
    result.append(np.ma.average(row, weights=w))
df['Total'] = result
df.head(3)
```
