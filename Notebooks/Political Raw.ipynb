```python
import pandas as pd
```

```python
col = ['State', 'ST', 'Fips', 'County',
    'Republicans 2016', 'Democrats 2016',
    'Republicans 2012','Democrats 2012', 
    'Republicans 2008', 'Democrats 2008']
df = pd.read_csv('../Data/Politics/formatted-usa-2016-presidential-election-by-county.csv', usecols=col)
df.head(3)
```

```python
df['2016']=pd.qcut(df['Republicans 2016'], 10, labels=[i for i in range(1,11)], duplicates='drop', )
df['2012']=pd.qcut(df['Republicans 2012'], 10, labels=[i for i in range(1,11)], duplicates='drop', )
df['2008']=pd.qcut(df['Republicans 2008'], 10, labels=[i for i in range(1,11)], duplicates='drop', )
df.head(3)
```

```python
df['party'] = [(row[-4]*10 + row[-3]*4 + row[-2]*3)/17 for i, row in df.iterrows()]
df.head(3)
```

```python
df2 = pd.DataFrame(df['Fips'])
df2.columns = ['FIPS']
df2['political_party'] = df['party']
df2.to_csv('../Data/Cleaned/Political_raw.csv')
```
