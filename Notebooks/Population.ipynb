```python
columns = ['Geographic Area', '2019', 'per_change']
df = pd.read_csv('../Data/Demographics/population_census_data.csv', encoding='iso-8859-1', usecols=columns )
df.head(4)
```

```python
df_map = pd.read_csv('../Data/Demographics/US_FIPS_Codes.csv', encoding='iso-8859-1')
df_map['m'] = df_map['County Name'] + ', ' + df_map['State']
df_map
```

```python
# df['County'] = [i.replace('.', '').split(' County,')[0].split(',')[0] for i in df['Geographic Area']]
# df['State'] = [i.split(',')[1] for i in df['Geographic Area']]
county = [i.replace('.', '').split(' County,')[0].split(',')[0] for i in df['Geographic Area']]
state = [i.split(',')[1] for i in df.iloc[:,0]]
df['m'] = pd.Series(county) + ', ' + pd.Series(state)
```

```python
df
```

```python
df_merged = pd.merge(df, df_map, left_index=True, right_index=True , on='m').drop(['m', 'Geographic Area'], axis=1)
df_merged
```

```python
df_merged.to_csv('../Data/Cleaned/Demographics_Population.csv')
```
